```python
def valid_password(password):
  upper = 0
  digit = 0
  special = 0
  has_min_length = 0

  if len(password) >= 8:
    has_min_length = 1

  for ch in password:
    if ch.isupper():
      upper = 1
    if ch.isdigit():
      digit = 1
    if not ch.isalnum(): # Fixed: checks if character is not alphanumeric
      special = 1

  total = upper + digit + special + has_min_length

  if total == 4:
    return "strong"
  elif total == 3:
    return "medium"
  else:
    return "weak"
```
