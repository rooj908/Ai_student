```python
menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

# Orders list, kuch items menu me nahi hain (jaise "cookei")
orders = [
    {"tea": 2, "cookei": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookei": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]
```

```python
processed_orders = []  # Final list jisme cleaned orders store honge

# Step 1: Har order ke liye loop
for order in orders:
    clean_order = {}  # Step 2: Har order ke liye naya empty dictionary

    # Step 3: Har item aur quantity ke liye loop
    for item in order:
        qty = order[item]  # Quantity nikal lo

        # Step 4: Check ke item menu me hai ya nahi
        if item in menu:

            # Step 5: Agar quantity inventory se zyada ho to adjust karo
            if qty > inventory[item]:
                qty = inventory[item]

            # Step 6: Clean order me add karo
            clean_order[item] = qty

    # Step 7: Processed orders me add karo
    processed_orders.append(clean_order)

# Step 8: Result print karo
print(processed_orders)
```
